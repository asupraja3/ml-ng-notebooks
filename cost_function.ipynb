```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample data: x = hours studied, y = scores
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 1.3, 3.75, 2.25])
m = len(x)
```

```python
# Cost function J(w, b)
def cost_function(x, y, w, b):
  cost = 0
  for i in range(m):
    f_wv_i = w * x[i] + b
    cost+= (f_wv_i - y[i])**2
  return cost/(2*m)
```

```python
# 1️⃣ Plot Cost vs w with b = 0
#generates 100 evenly spaced values between -2 and 4
w_vals = np.linspace(-2, 4, 100)
cost_vals = [cost_function(x, y, w, 0) for w in w_vals]

plt.figure(figsize=(8, 5))
plt.plot(w_vals, cost_vals, label='J(w) at b=0')
plt.xlabel('Weight (w)')
plt.ylabel('Cost J(w, b=0)')
plt.title('Cost vs Weight (b=0)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 2️⃣ Plot 3D Cost Surface: J(w, b)
from mpl_toolkits.mplot3d import Axes3D

W = np.linspace(-2, 4, 50)
B = np.linspace(-2, 2, 50)
W, B = np.meshgrid(W, B)
Z = np.zeros_like(W)

# Compute cost surface
for i in range(W.shape[0]):
    for j in range(W.shape[1]):
        Z[i, j] = cost_function(x, y, W[i, j], B[i, j])

# 3D Plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W, B, Z, cmap='viridis', alpha=0.9)
ax.set_xlabel('Weight (w)')
ax.set_ylabel('Bias (b)')
ax.set_zlabel('Cost J(w, b)')
ax.set_title('3D Cost Function Surface')
plt.show()

```

```python
#3️⃣ Contour Plot of Cost Function
plt.figure(figsize=(8, 6))
cp = plt.contour(W, B, Z, levels=30, cmap='viridis')
plt.clabel(cp, inline=True, fontsize=8)
plt.xlabel('Weight (w)')
plt.ylabel('Bias (b)')
plt.title('Contour Plot of Cost Function J(w, b)')
plt.grid(True)
plt.show()
```
